```julia
using Distributions
using Plots

f = 1.7
x = Beta(f,f)
r = 0:0.02:1
d = diff(cdf(x, r))
scaled_data = d * 0.93/maximum(d)

# plot(d * 1/maximum(d))

rounded = [round(x; digits=2) for x in scaled_data]

rounded[1] = rounded[50] = 0.1
rounded[28] = 0.93
rounded[38] = 0.77

out = vcat(zeros(28), rounded, zeros(18))
println(out)

plot(out)
```

```julia
using Distributions
using Plots

x = Beta(2,3.5)
r = 0:0.01:1
d = diff(cdf(x, r))
scaled_data = d[1:96] * 0.9/maximum(d)

# plot(d * 1/maximum(d))

rounded = [round(x; digits=2) for x in scaled_data]

rounded[28] = 0.9 
rounded[47] = 0.7
rounded[65] = 0.35
rounded[80] = 0.1

println(rounded)


plot(rounded)
```
